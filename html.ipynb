```python
import datasets

dataset = datasets.load_dataset("philschmid/markdown-documentation-transformers")

dataset

df = dataset["train"].to_pandas()

df.head()
```

```python
from docling.document_converter import DocumentConverter
from docling.document_converter import DocumentConverter, DocumentStream
from io import BytesIO

markdown = df.sample(1)["html"].iloc[0]

# Docling conversion
converter = DocumentConverter()

# Convert document
buf = BytesIO(markdown.encode('utf-8'))
source = DocumentStream(name="my_doc.html", stream=buf)
doc = converter.convert(source)
text = doc.document.export_to_text()

with open("html.txt", "w") as f:
    f.write(text)
```

```python
from bs4 import BeautifulSoup

markdown = df.sample(1)["html"].iloc[0]

html_content = markdown
soup = BeautifulSoup(html_content, 'html.parser')
text = soup.get_text()

with open("html.txt", "w") as f:
	f.write(text)
```
